```python
import folium

# Create a map centered around the provided coordinates
latitude = 46.79152
longitude = 4.77815

# Create a map object
mymap = folium.Map(location=[latitude, longitude], zoom_start=15)

# Add a marker to the map for the given location
folium.Marker([latitude, longitude], popup="Lieu Dit Le Roussot, 2 Chemin du Moulin").add_to(mymap)

# Display the map
mymap
```
